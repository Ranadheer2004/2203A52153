```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
b = pd.read_csv('/content/Salary_Data.csv')
m=b["YearsExperience"]
y=b['Salary']
plt.scatter(m,y)
m_w = np.column_stack((np.ones(len(m)), m))
print(m)
print(m_w)
plt.scatter(m, y)
plt.show()
m_w_t = m_w.T
#print("\nTranspose of m with ones:")
#print(m_w_t)
m_w_tintoy = np.dot(m_w_t, y)
#print(m_w_tintoy)
m_wintox_w_t = np.dot(m_w_t, m_w)
#print(x_wintox_w_t)
m_wintox_w_t_inv = np.linalg.inv(m_wintox_w_t)
#print(m_wintox_w_t_inv)
weights = np.dot(m_wintox_w_t_inv, m_w_tintoy)
#print("\nweights=")
print(weights)
y_p = np.dot(m_w, weights)
print("\nPredicted y values:")
print(y_p)
intercept = weights[0]
slope = weights[1]
print("\nBest-fit line equation:")
print(f"y = {slope:.2f} * m + {intercept:.2f}")
line_eq = f"y = {slope:.2f} * m + {intercept:.2f}"
m_fit = np.linspace(min(m), max(m), 100)
y_fit = slope * m_fit + intercept
plt.plot(m_fit, y_fit, color='red', label=line_eq)
plt.legend()
plt.scatter(m,y)
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Best-Fit Line and Data Points')
plt.show()
```
